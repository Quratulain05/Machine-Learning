```python
import pandas as pd

# Correct way to read a CSV file
df = pd.read_csv("score.csv")  # If file is in same folder
df
```

```python
import matplotlib.pyplot as plt
plt.scatter(df["Hours"], df["Scores"])
plt.xlabel("Hours Studied")
plt.ylabel("Exam Score")

plt.show()
```

```python
X = df[["Hours"]]
y = df["Scores"]
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X, y)
```

```python
import pandas as pd
# Create a DataFrame with the same feature name used in training
new_data = pd.DataFrame([[6]], columns=["Hours"])  # "Hours" must match training data
prediction = model.predict(new_data)
print(f"Predicted score for 6 hours: {prediction[0]:.2f}")
```

```python
print(f"Slope (coefficient): {model.coef_[0]}")
print(f"Intercept: {model.intercept_}")
```

```python
from sklearn.metrics import r2_score
y_pred = model.predict(X)
print(f"R² Score: {r2_score(y, y_pred):.2f}")  # 1.0 = perfect fit
```
